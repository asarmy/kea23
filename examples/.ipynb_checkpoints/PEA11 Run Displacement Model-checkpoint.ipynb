## Loop for multiple scenarios

```python
# User-defined cases
magnitude, location, percentile, submodel = (
    [6.8, 7.4],
    [0.25, 0.5],
    [0.5, -1],
    ["elliptical", "bilinear", "quadratic"],
)

# Define all scenarios
if not isinstance(magnitude, list):
    magnitude = [magnitude]
if not isinstance(location, list):
    location = [location]
if not isinstance(percentile, list):
    percentile = [percentile]
if not isinstance(submodel, list):
    submodel = [submodel]

# Perform calculations
scenarios = list(product(magnitude, location, percentile, submodel))
df = pd.DataFrame()
for scenario in scenarios:
    results = run_displacement_model.run_model(
        magnitude=scenario[0],
        location=scenario[1],
        percentile=scenario[2],
        submodel=scenario[3],
    )
    df = pd.concat([df, results])

# Clean up variables
del magnitude, location, percentile, submodel

df
```

# Calculate slip profiles

```python
# User-defined cases
location = np.arange(0, 1.01, 0.01).tolist()
magnitude = 7
percentile = 0.84
submodel = ["elliptical", "bilinear", "quadratic"]

# Define all scenarios
if not isinstance(magnitude, list):
    magnitude = [magnitude]
if not isinstance(location, list):
    location = [location]
if not isinstance(percentile, list):
    percentile = [percentile]
if not isinstance(submodel, list):
    submodel = [submodel]

# Perform calculations
scenarios = list(product(magnitude, location, percentile, submodel))
df = pd.DataFrame()
for scenario in scenarios:
    results = run_displacement_model.run_model(
        magnitude=scenario[0],
        location=scenario[1],
        percentile=scenario[2],
        submodel=scenario[3],
    )
    df = pd.concat([df, results])

df
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 4))

for shape in ["elliptical", "bilinear", "quadratic"]:
    _df = df[df["model_name"] == shape]
    _df = _df.sort_values(by="location", ascending=True)
    x, y = _df["location"], _df["displ"]
    ax.plot(x, y, label=shape)

title = f"Magnitude {magnitude}, Percentile={percentile}"
ax.set(xlabel="x/L", ylabel="Principal Displacement (m)", title=title)
ax.legend(loc="best")

# Clean up variables
del magnitude, location, percentile, submodel
```
